```python
number=int(input("enter the number"))
factor_count=0
i=1
while i<= number:

  if number % i == 0:
    factor_count = factor_count + 1

  i=i+1

if factor_count == 2:
    print("prime number")
else:
    print("not a prime number")
```

```python
#A prime number has only two factors: 1 and itself.
#We use a loop to check each number from 1 to the given number.
# Whenever the remainder is 0, we count it as a factor.
#If the factor count is 2, the number is prime.
```
