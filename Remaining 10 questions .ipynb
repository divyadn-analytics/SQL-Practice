```python
# 15. ATM withdrawal validation

balance = int(input("Enter your account balance: "))
withdraw = int(input("Enter withdrawal amount: "))
min_balance = 500   # minimum balance required

# Condition 1: Must be multiples of 100
if withdraw % 100 != 0:    (#% 100 != 0 → not a multiple of 100, % → remainder,!= → not equal)
    print("Withdrawal amount must be in multiples of 100")

# Condition 2: Cannot withdraw more than balance
elif withdraw > balance:
    print("Insufficient balance")

# Condition 3: Must keep minimum balance
elif balance - withdraw < min_balance:
    print("Cannot withdraw: minimum balance of 500 must remain")

# If all conditions pass
else:
    balance -= withdraw
    print("Withdrawal successful!")
    print("Remaining balance:", balance)
```

```python
#16.Shopping discount calculator based on purchase amount and membership
amount = int(input("Enter purchase amount: $"))
member = input("Do you have membership? (yes/no): ").lower()

if member == "yes":
    discount = 0.10   # 10% discount for members
else:
    discount = 0.05   # 5% discount for non-members

final_price = amount - (amount * discount)

print("Discount applied:", discount * 100, "%")
print("Final price after discount: $", final_price)
```
