```python
cosmetics = {"perfume": 150, "face cream": 70, "eyeliner": 80}
electronics = {"mobile phone": 500, "laptop": 1000, "smart watch": 150}
fabric = {"t-shirt": 20, "jeans": 45, "jacket": 80, "shoes": 50}
home_app = {"mixer grinder": 150, "television": 85, "electric ricecooker": 75}

cart = []

while True:
    print("\n----- Welcome to Shopping Mall -----")
    print("1. Cosmetics")
    print("2. Electronics")
    print("3. Fabric")
    print("4. Home appliances")
    print("5. Exit and print final bill")

    option = input("Choose a category: ")

    if option == "1":
        product = cosmetics
    elif option == "2":
        product = electronics
    elif option == "3":
        product = fabric
    elif option == "4":
        product = home_app
    elif option == "5":
        break
    else:
        print("Invalid option. Try again.")
        continue

    print("\nAvailable products:")
    for name, price in product.items():
        print(name, "-", price)

    choice = input("Enter the product: ").lower()

    if choice in product:
        quantity = int(input("Enter quantity: "))
        total = product[choice] * quantity

        cart.append([choice, quantity, total])
        print("Added to cart!")
    else:
        print("Sorry! Product is not available in this category.")

print("\n------ Final Bill ------")
grand_total = 0

for item in cart:
    print(item[0], " Quantity:", item[1], " Total:", item[2])
    grand_total += item[2]

print("Total amount:", grand_total)
```
