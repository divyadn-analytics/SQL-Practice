```python
product={"apple":30,"banana":10,"milk":20,"yogurt":15}
cart=[]
while True:
    choice=input("enter the product:").lower()
    quantity=int(input("enter the quantity:"))
    if choice in product:
        print("product is available")
        print("price:",product[choice])
        total=product[choice]*quantity
        print("total:",total)
        cart.append([choice,quantity,total])
        print(cart)
    else:
        print("sorry!product is not available")
    again=input("do you want to continue shooping?").lower()
    if again=='no':
       break
print("------Finalbill----------")
grand_total=0
for items in cart:
    print(items[0],items[1],items[2])
    grand_total=grand_total+items[2]
print("total amount:",grand_total)
```
