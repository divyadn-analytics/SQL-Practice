```python
import math # its a math toolbox ,advance math fn stored in a special library called math module.
# build a calculator adding power and square root
print("basic calculator")
print("-----------------")
try:
    print("choose opertaions")
    print("1. addition")
    print("2. subtraction")
    print("3. multiplication")
    print("4. division")
    print("5. power")
    print("6. square root")

    choice = input("enter choice (1-6):")

    num1=float(input("enter first number: "))

    if choice != '6': # != not equal to (# ask for num2 only if choice is not square root)
        num2=float(input("enter second number: "))

    if choice == '1':
        print("result:", num1 + num2)
    elif choice == '2':
        print("result:", num1 - num2)
    elif choice =='3':
        print("result:", num1 * num2)
    elif choice == '4':
        if num2 == 0:
            print("error:division by zero is not allowed.")
        else:
            print("result:", num1 / num2)
    elif choice == '5':
        print("result.",num1 ** num2)
    elif choice == '6':
        if num2 < 0: #cannot take square root of negative numbers in normal math.
            print("error : cannot take square root of a negetive number.") # If the number is negative, the program shows an error message instead of trying to calculate the square root.
        else:
            print("result:",math.sqrt(num1)) # perform the function using math. square root

    else:
        print("error:invalid opertaion. choose between 1-6.")

except ValueError:  # (ValueError happens when Python expects a number but receives something else like abc.)
    print("error: please enter numeric values only.") # (exception handling program shows only numeric which is useful for user)


except Exception as e:  #(exception is the error catcher anything otherthan value error)
    print("unexpected error:", str(e))  #(it stores actual error message inside variable e ,helps us understand what went wrong/ shows real error)
```
